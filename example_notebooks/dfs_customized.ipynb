```python
from webpagelinkminer.link_extractor import WebPageLinkExtractor
```

```python
# Create an instance of WebPageLinkExtractor with custom parameters
extractor = WebPageLinkExtractor(
    base_url='https://python.langchain.com/en/latest/',
    timeout=15,
    max_depth=5,
    parser='html.parser',
    algorithm='dfs'
)
```

```python
# Get all the links using the DFS algorithm and custom parameters
links = extractor.get_links()
```
